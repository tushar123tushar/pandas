```python
import pandas as pd
```

```python
titanic  = pd.read_csv('Titanic-Dataset.csv')
titanic
```

```python
titanic['Sex'] = titanic['Sex'].map({"male":0 , "female" : 1})
```

```python
titanic
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age']).mean()
```

```python
titanic
```

```python
titanic.drop(['Ticket' , 'Cabin'], axis=1 , inplace = True)
```

```python
titanic
```

```python
titanic['Embarked'].value_counts()
```

```python
titanic['Name']
```

```python
titanic.drop(['Name' , 'Embarked' ], axis=1 , inplace=True)
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic
```

```python
new_data = titanic.groupby('Sex')['Survived'].mean()
print(new_data)
```

```python
new_Data = titanic.groupby('Pclass')['Survived'].mean()
print(new_Data)
```

```python
import seaborn as sns
import  matplotlib.pyplot as plt
```

```python
sns.histplot(data=titanic, x='Age', hue='Survived', kde=True)
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
correlation = titanic.corr()
print(correlation)
sns.heatmap(correlation , annot=True , cmap='coolwarm', fmt='.3f')
plt.title("Find The Correlation Find")
plt.show()
```

```python
sns.histplot(titanic['Age'], kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=titanic)
plt.title('Survival Rate by Pclass')
plt.show()
```
